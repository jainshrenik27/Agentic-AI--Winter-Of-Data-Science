```python
from transformers import pipeline

summarizer = pipeline(
    "summarization",
    model="facebook/bart-large-cnn"
)

text = """
Artificial Intelligence has seen rapid growth over the past decade, transforming
industries such as healthcare, finance, education, and transportation. Machine
learning models can now analyze massive datasets and identify patterns that would
be impossible for humans to detect manually. Despite  these advances, AI systems
still require careful design, ethical considerations, and continuous monitoring
to avoid bias and misuse. Researchers are actively working on making AI models
more interpretable and trustworthy. As AI continues to evolve, collaboration
between technologists, policymakers, and society will be crucial.
"""

# Print original text length
original_word_count = len(text.split())
original_char_count = len(text)

print("Original Text:")
print(text)
print(f"\nOriginal Length: {original_word_count} words, {original_char_count} characters")

# Generate summary
summary_output = summarizer(
    text,
    min_length=40,
    max_length=80,
    do_sample=False
)

summary_text = summary_output[0]['summary_text']
print(summary_text)
print(f'Length of summary text is {len(summary_text)} characters')
```

```python
from transformers import pipeline

# Create a text-generation pipeline
generator = pipeline(
    "text-generation",
    model="gpt2"
)

# Prompt
prompt = "In 2030, AI systems will"

# Generate two continuations
outputs = generator(
    prompt,
    num_return_sequences=2,
    max_new_tokens=50,
    do_sample=True
)

# Print results
print("PROMPT:")
print(prompt)

for i, output in enumerate(outputs, start=1):
    print(f"\n--- Generated Continuation {i} ---")
    print(output["generated_text"])
```

```python
from transformers import pipeline

# Create a sentiment analysis pipeline
sentiment_analyzer = pipeline("sentiment-analysis")

# List of movie reviews
reviews = [
    "The movie was absolutely fantastic and kept me engaged throughout.",
    "I found the film boring and far too long.",
    "The acting was decent, but the story was predictable.",
    "An amazing cinematic experience with stunning visuals.",
    "The plot made no sense and the ending was disappointing."
]

# Run sentiment analysis on all reviews at once
results = sentiment_analyzer(reviews)

# Print results in a readable format
for review, result in zip(reviews, results):
    print(f"\nReview: {review}")
    print(f"Sentiment: {result['label']}, Score: {result['score']}")
```
